```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, cut_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/abc-csv/abc.csv")
df.head()
```

```python
df1 = df[['Fare', 'Age', 'PassengerId']]
df1.head()
```

```python
df1.set_index("PassengerId", inplace=True)
```

```python
df1.head()
```

```python
df1['Fare'].isnull().sum()
```

```python
b = df1['Age'].mean()
s = round(b, 2)
s
```

```python
df1['Age'].fillna(s, inplace=True)
```

```python
df1['Age'].isnull().sum()
```

```python
data = df1[['Fare', 'Age']]
data
```

```python
single_linkage = linkage(data, method='single')
single_linkage
```

```python
complete_linkage = linkage(data, method='complete')
complete_linkage
```

```python
num_clusters = 10
```

```python
cluster_assignments = cut_tree(single_linkage, n_clusters=num_clusters).flatten()
cluster_assignments
```

```python
print("Cluster Assignments:")
for i, assignment in enumerate(cluster_assignments):
    print(f"Data point {i+1}: Cluster {assignment}")
```
