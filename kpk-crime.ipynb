```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/kpk-crime-statisticscsv/kpk-crime-statistics.csv')
df.head()
```

```python
columns = ['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11']
df.drop(columns=columns, inplace = True)

df.columns
```

```python
df.head()
```

```python
df['crime type'].value_counts()
```

```python
peshawar_data = df[df["district"] == "Peshawar"]

# Group data by crime type and sum the cases
crime_type_counts = peshawar_data.groupby("crime type")["cases"].sum()

# Create a bar chart
plt.figure(figsize=(10, 6))
crime_type_counts.plot(kind="bar")
plt.title("Crime Types in Peshawar District")
plt.xlabel("Crime Type")
plt.ylabel("Number of Cases")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (16, 9))
chart = df.groupby('crime type')['cases'].sum()
plt.pie(chart, labels=chart.index, autopct='%1.1f%%', startangle=140)
plt.title('Crimes types from 2011 to 2015 in Kpk')
plt.show();
```

```python
df.groupby('district')['cases'].sum().plot(kind='bar')
plt.ylabel('Districts')
plt.xlabel('Number of cases')
plt.title('Number of cases per district')
plt.show();
```

```python
df.groupby('division')['cases'].sum().plot(kind='barh')
plt.ylabel('Divisions')
plt.xlabel('Number of cases')
plt.title('Number of cases per division')
plt.show();
```

```python
yearly_cases = df.groupby('year')['cases'].sum()
plt.bar(yearly_cases.index, yearly_cases.values)
plt.xlabel('years')
plt.ylabel('Number of Cases')
plt.title('Number of Cases Per Year')
plt.show();
```
